```python
import numpy as np
from scipy.special import spence
import matplotlib.pyplot as plt
```

```python
mp = .938272081
me = 0.5109989461 * 0.001
alpha = 1/137.036
```

```python
Ee, Q2, Pe, thetae, thetap =  np.loadtxt("kinematics.csv", delimiter=",").T
```

```python
delta_vertex = alpha/np.pi * (1.5 * np.log(Q2/me**2) -2 - 0.5*np.log(Q2/me**2)**2 + np.pi**2/6)
delta_vac    = alpha/np.pi * (2./3.) * (-5./3. + np.log(Q2/me**2))
```

```python
#Note that these are the elastic scattering corrections
Eeprime      = (mp*Ee - Q2/2) / mp
Eetilde      = Eeprime
Eeprime_el   = 0.01 * Ee + Eeprime
Eeprimetilde = Ee
eta          = Ee / Eeprime_el
deltaEs      = eta * (Eeprime_el - Eeprime)


delta_r1     = alpha/np.pi * np.log(deltaEs**2 /Eetilde/Eeprimetilde) * (np.log(Q2/me**2)-1)

delta_r2 = alpha/np.pi* (-0.5* (np.log(Eetilde/Eeprimetilde)**2) + 0.5 * np.log(Q2/me**2)**2 - np.pi**2/3 + spence(1 - np.cos(thetae/2)**2))
```

```python
rho = np.sqrt(Q2 + 4*mp**2)
aks = (np.sqrt(Q2) + rho)**2/4/mp**2

Eprime_N = mp + Ee - Eeprime
pprime_N = np.sqrt(Eprime_N ** 2 - mp ** 2)
delta_1_reproduced   = 2*alpha/np.pi * (np.log(4*(deltaEs)**2/Q2/aks) * np.log(eta) + spence(1-(1-(eta/aks))) - spence(1-(1 - 1/eta/aks)) ) 

delta_2_reproduced   = alpha/np.pi * ( np.log(4*(deltaEs)**2/mp**2)*(Eprime_N/pprime_N*np.log(aks)-1) + 1 
                        + Eprime_N/pprime_N * (-0.5*np.log(aks)**2 - np.log(aks)*np.log(rho**2/mp**2) + np.log(aks)
                        - spence( 1 - (1-1/aks/aks)) +2*spence(1-(-1/aks)) + np.pi**2/6 ) ) #A77
```
